```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# 1. Load your CSV data
data = pd.read_csv(r'D:\Projects\Investment _Prediction\Stock_data\AAPL.csv') # Assuming it contains 'Open', 'High', 'Low', 'Close'
```

```python
# 2. Feature Engineering
data['Open-Close'] = data['Open'] - data['Close']
data['High-Low'] = data['High'] - data['Low']

# 3. Features and Target
X = data[['Open', 'High', 'Low', 'Open-Close', 'High-Low']]  # Features
y = data['Close'].shift(-1).fillna(method='ffill')  # Target: Predict next day's Close
```

```python
# 4. Split data into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# 5. Normalize the features (optional but recommended for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

### 6. KNN Regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
```

```python

# 7. Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python

# 8. Evaluate Models
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print(f"KNN Regressor - Mean Squared Error: {mse_knn}, R-squared: {r2_knn}")
print(f"Random Forest - Mean Squared Error: {mse_rf}, R-squared: {r2_rf}")
```

```python

# 9. Plotting Actual vs Predicted for KNN and Random Forest
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(y_test.values, label='Actual Close Price', color='blue')
plt.plot(y_pred_knn, label='KNN Predicted Close Price', color='red')
plt.title('KNN Regressor: Actual vs Predicted')
plt.xlabel('Test Sample Index')
plt.ylabel('Close Price')
plt.legend()
```

```python

plt.subplot(1, 2, 2)
plt.plot(y_test.values, label='Actual Close Price', color='blue')
plt.plot(y_pred_rf, label='Random Forest Predicted Close Price', color='green')
plt.title('Random Forest: Actual vs Predicted')
plt.xlabel('Test Sample Index')
plt.ylabel('Close Price')
plt.legend()

plt.tight_layout()
plt.show()

```
